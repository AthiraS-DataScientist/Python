8. Program to validate vehicle registration numbers

An Indian vehicle registration number generally has:

* 2 uppercase letters (state code)
* followed by 2 digits (district code)
* followed by 2 uppercase letters (unique series)
* ends with 4 digits (unique vehicle number)

```python
vehicle_num="KA01AB1234"
pattern="^[A-Z]{2}\d{2}[A-Z]{2}\d{4}"

result=re.match(pattern,vehicle_num)

if result!=None:
  print("Valid vehicle number")
else:
  print("Invalid vehicle number")
```

# Collections Module

This module implements specialized
container data types providing  alternatives
to Python's general purpose built-in
containers like dict, list, set, and tuple.

 Some Popular containers

collections.Counter

A Counter is a container that tracks how
many times equivalent values are added.


```python
import collections as c

my_list=[1,2,3,3,4,2,8]

frequency=c.Counter(my_list)

print(frequency)
```

```python
string="Cranes Varsity Private Limited"

frequency=c.Counter(string)

print(frequency)
```

Application of Counter:

Migratory Bird

 You have been asked to help study the population
of birds migrating across the continent.

 Each type of bird you are interested in will be
identified by an integer value.

 Each time a particular kind of bird is spotted, its id
number will be added to your array of sightings.

 You would like to be able to find out which type of
bird is most common given a list of sightings.

Counter - most_common method

 most_common(n=None)

  List the n most common elements and their counts from
the most common to the least.  

 If n is None, then list all element counts.

  Returns a list of n elements, where each element is a
tuple.

```python
data = "KARNATAKA"
frequency = Counter(data)
frequency
```

```python
frequency.most_common(1)# character with highest frequency
```

```python
frequency.most_common(2)# TOP 2 most occuring characters
```

```python
frequency.most_common() # frequency of all characters in decreasing order of frequency
```

```python
my_list=frequency.most_common()
my_list.reverse()
my_list
```
