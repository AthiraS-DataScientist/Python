#### Understanding Exceptions

Exception handling in Python allows you to handle errors gracefully and take corrective actions without stopping the execution of the program.

##### What Are Exceptions?
Exceptions are events that disrupt the normal flow of a program. They occur when an error is encountered during program execution. Common exceptions include:

- ZeroDivisionError: Dividing by zero.
- FileNotFoundError: File not found.
- ValueError: Invalid value.
- TypeError: Invalid type.

```python
a=b
```

```python
try:
  a=b

except:
  print("Exception occured")
```

```python
try:
  a=c

except NameError as ex:
  print (ex)
```

#ZeroDivisionError: Dividing by zero

```python
1/0
```

```python
try:
  result=1/0

except NameError as ex:
  print (ex)
except ZeroDivisionError as ab:
  print (ab)
```

```python
try:
  num= int(input("Enter a number: "))
  result=1/num

except ZeroDivisionError as ex:
  print(ex)

else:
  print(result)
```

```python
## try,except,else and finally
try:
    num = int(input("Enter a number: "))
    result = 10 / num
    a=b
except NameError as ex:
    print("That's not a valid name!")
except ZeroDivisionError:
    print("You can't divide by zero!")

else:
    print(f"The result is {result}")
finally:
    print("Execution complete.")
```

#Type Error:

It occurs when an operation is performed on an object of inappropriate type.


Common causes:

1. Incorrect data type

2. Passing a wrong argument to function

3. Attempting operations between incompatible types

```python
def add(a,b):
  try:
    return a+b
  except TypeError:
    print("Invalid value")

add(1,"fh")
```

```python
try:
  b=1+2+"abc"

except TypeError:
  print("Invalid value")
```

```python
def divide(a,b):
  try:
    return a/b
  except ZeroDivisionError:
    print("Zero Division Error")
  except TypeError:
    print("Type Error")

divide(1,0)
```

#Collections Module

It implements specialized container data types providing an alternative to python's general dtypes

Some popular Containers are:

1. Counter

2. Named Tuple

3. Ordered Dict

4. deque


```python
import collections as c

b=[1,1,1,1,5,8,9,6,6]

s=c.Counter(b)
s
```

```python
d="Cranes Varsity"

e=c.Counter(d)
```

most_common method

```python
y=[1,5,8,9,6,6,6,6,8,8,1]
z=c.Counter(y)
z.most_common(1) #character with the highest frequency
```

```python
z.most_common(2) #top 2 most occurong characters
```

```python
z.most_common() #frequency of all characters in a decreasing order
```
